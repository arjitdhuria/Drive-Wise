```python
import pandas as pd
import numpy as np

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
cars_data=pd.read_csv('Cardetails.csv')
```

```python
cars_data.drop(columns=['torque'],inplace=True)
```

```python
# DATA ANALYSIS
```

```python
def getBrand(car_name):
    car_name=car_name.split(' ')[0]
    return car_name.strip(' ')
```

```python
def extract_float(value):
    try:
        return float(str(value).split()[0])
    except:
        return None 
```

```python
getBrand('Maruti swift dzire')
```

```python
cars_data['name']=cars_data['name'].apply(getBrand)
```

```python
cars_data
```

```python
cars_data['mileage'] = cars_data['mileage'].apply(extract_float).astype('float64')
cars_data['max_power'] = cars_data['max_power'].apply(extract_float).astype('float64')
cars_data['engine'] = cars_data['engine'].apply(extract_float).astype('float64')
```

```python
cars_data.head()
```

```python
cars_data['name'] = cars_data['name'].replace(
    ['Maruti', 'Skoda', 'Honda', 'Hyundai', 'Toyota', 'Ford', 'Renault',
     'Mahindra', 'Tata', 'Chevrolet', 'Datsun', 'Jeep', 'Mercedes-Benz',
     'Mitsubishi', 'Audi', 'Volkswagen', 'BMW', 'Nissan', 'Lexus',
     'Jaguar', 'Land', 'MG', 'Volvo', 'Daewoo', 'Kia', 'Fiat', 'Force',
     'Ambassador', 'Ashok', 'Isuzu', 'Opel'],
    list(range(1, 32))  # 1 to 31
)
```

```python
cars_data['transmission'] = cars_data['transmission'].replace(['Manual', 'Automatic'], [1, 2])
```

```python
cars_data
```

```python
cars_data['seller_type'] = cars_data['seller_type'].replace(
    ['Individual', 'Dealer', 'Trustmark Dealer'], [1, 2, 3]
).astype(int)
```

```python
cars_data['fuel'] = cars_data['fuel'].replace(
    ['Diesel', 'Petrol', 'LPG', 'CNG'],
    [1, 2, 3, 4]
).astype(int)
```

```python
cars_data.info()
```

```python
cars_data
```

```python
cars_data.reset_index(inplace=True)
```

```python
cars_data
```

```python
cars_data['owner'] = cars_data['owner'].replace(
    ['First Owner', 'Second Owner', 'Third Owner',
     'Fourth & Above Owner', 'Test Drive Car'],
    [1, 2, 3, 4, 5]
).astype(int)
```

```python
cars_data
```

```python
cars_data.drop(columns=['index'], inplace=True)
```

```python
input_data = cars_data.drop(columns=['selling_price'])
output_data =cars_data['selling_price']
```

```python
x_train, x_test, y_train, y_test = train_test_split(input_data, output_data, test_size=0.2)
```

```python
model = LinearRegression()
x_train = x_train.dropna()
y_train = y_train.loc[x_train.index]
model.fit(x_train, y_train)
```

```python
x_train
```

```python
predict = model.predict(x_test)
```
